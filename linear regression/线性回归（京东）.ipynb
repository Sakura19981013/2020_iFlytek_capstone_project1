```python
import pandas as pd
import numpy as np
df = pd.read_csv('jd_after529.csv')
df
```

```python
x_var=df2.loc[:,['是否为展位广告','是否为触点广告','是否为海投广告','是否为快车广告','是否为京东直投','总费用']]
```

**点击率/转化率**

```python
y_var=df2.loc[:,['点击率/转化率']]
```

```python
import statsmodels.formula.api as smf
data=pd.concat([x_var,y_var],axis=1)
data
```

```python
mod = smf.ols(formula='Q("点击率/转化率")~是否为触点广告+是否为海投广告+是否为快车广告+是否为京东直投+总费用',data=data)#展位广告位基准
```

```python
res = mod.fit()
print(res.summary())
```
